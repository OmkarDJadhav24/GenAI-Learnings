```python
import spacy

# Load medium-sized model with word vectors
nlp = spacy.load("en_core_web_md")

# Compare similarities
word1 = nlp("king")
word2 = nlp("queen")
word3 = nlp("apple")

print("Similarity between king and queen:", word1.similarity(word2))
print("Similarity between king and apple:", word1.similarity(word3))
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

words = ["king", "queen", "man", "woman", "apple", "orange"]
vectors = [nlp(word).vector for word in words]

pca = PCA(n_components=2)
reduced = pca.fit_transform(vectors)

plt.figure(figsize=(8, 6))
for i, word in enumerate(words):
    plt.scatter(reduced[i][0], reduced[i][1])
    plt.annotate(word, (reduced[i][0]+0.01, reduced[i][1]+0.01))
plt.title("Word Embeddings in 2D Space")
plt.show()
```
